```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
```

```python
df.head()
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
plt.xlabel=('sepal_lenght')
plt.ylabel=('sepal_width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='red',marker='D',label='setosa')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='o',label='versicolor')
plt.legend()
plt.show()
```

```python
plt.xlabel=('petal_lenght')
plt.ylabel=('petal_width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='D',label='setosa')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='o',label='versicolor')
plt.legend()
plt.show()
```

```python
x=df.drop(['target','flower_name'],axis=1)
y=df.target
```

```python
from sklearn.model_selection import train_test_split
X_TR,X_T,Y_TR,Y_T=train_test_split(x,y,test_size=0.2, random_state=50)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_TR,Y_TR)
```

```python
knn.score(X_T,Y_T)
```

```python
y_pred=knn.predict(X_T)
```
